```python
import seaborn as sns
```

```python
plt.subplot(1,3,1)
sns.countplot(x = df['Road_surface_type'])
plt.subplot(1,3,2)
sns.countplot(x = df['Weather_conditions'])
plt.subplot(1,3,3)
sns.countplot(x = df['Light_conditions'])
```

```python
fig, axes = plt.subplots(3 , 1 , figsize = (12 , 15))
sns.countplot(x = df['Road_surface_type'] , ax = axes[0])
sns.countplot(x = df['Weather_conditions'] , ax = axes[1])
sns.countplot(x = df['Light_conditions'] , ax = axes[2])
plt.tight_layout()
plt.show()
```

```python
accident_hotspots = df.groupby(['Road_surface_type' , 'Weather_conditions' , 'Light_conditions']).size().unstack()
plt.figure(figsize = (14,12))
sns.heatmap(accident_hotspots , cmap = 'viridis' , annot = True , fmt = 'g' , cbar_kws = {'label' : 'Number of Accidents'})
plt.show()
```

```python
plt.figure(figsize = (14,6))
sns.countplot( x = 'Road_surface_type' , hue = 'Accident_severity' , data = df)
```

```python
plt.figure(figsize = (14,6))
sns.countplot(x = 'Weather_conditions' , hue = 'Accident_severity' , data = df)
```
